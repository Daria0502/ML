```python
# загрузка пакетов: инструменты ---
# работа с массивами
import numpy as np
# фреймы данных
import pandas as pd
# графики
import matplotlib as mpl
import matplotlib.pyplot as plt
# стили и шаблоны графиков на основе matplotlib
import seaborn as sns
# перекодировка символьных показателей
from sklearn.preprocessing import LabelEncoder, StandardScaler
# для таймера
import time

# загрузка пакетов: модели ---
# SVM
from sklearn.svm import SVC
# логистическая регрессия
from sklearn.linear_model import LogisticRegression
# метод главных компонент
from sklearn.decomposition import PCA
# конвейеры
from sklearn.pipeline import make_pipeline
# перекрёстная проверка и метод проверочной выборки
from sklearn.model_selection import cross_val_score, train_test_split
# для перекрёстной проверки и сеточного поиска
from sklearn.model_selection import KFold, GridSearchCV
# сводка по точности классификации
from sklearn.metrics import classification_report

# константы
# ядро для генератора случайных чисел
my_seed = 17  # номер варианта

# настройка стиля и отображения графиков - ИСПРАВЛЕННЫЕ СТИЛИ
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")

# Посмотрим доступные стили
print("Доступные стили matplotlib:")
print(plt.style.available)
```

```python
# загружаем данные
url = "https://raw.githubusercontent.com/ania607/ML/main/data/default_of_credit_card_clients.csv"
DF_raw = pd.read_csv(url, skiprows=0)  

# выясняем размерность фрейма
print('Число строк и столбцов в наборе данных:\n', DF_raw.shape)
```

```python
# первые строки
DF_raw.head()
```

```python
# информация о данных
print("Информация о столбцах:")
print(DF_raw.info())
```

```python
# исключаем ненужные столбцы
columns_to_drop = ['PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 
                   'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6', 'Y']

# Проверяем, какие столбцы действительно существуют в данных
existing_columns = [col for col in columns_to_drop if col in DF_raw.columns]
print("Столбцы для удаления:", existing_columns)

DF_processed = DF_raw.drop(columns=existing_columns, errors='ignore')

# зависимая переменная Y
y = DF_raw['Y']

# проверяем оставшиеся столбцы
print("\nОставшиеся столбцы:", DF_processed.columns.tolist())
print(f"Размерность признаков: {DF_processed.shape}")
print(f"Размерность целевой переменной: {y.shape}")
```

```python
# Проверяем распределение целевой переменной
print("Распределение целевой переменной Y:")
print(y.value_counts())
print("\nДоли:")
print(y.value_counts(normalize=True))
```

```python
# разделяем данные на обучающую и отложенную выборку (85% / 15%)
X_train, X_predict, y_train, y_predict = train_test_split(
    DF_processed, y, test_size=0.15, random_state=my_seed, stratify=y
)

print(f"Обучающая выборка: {X_train.shape}")
print(f"Отложенная выборка: {X_predict.shape}")
print(f"Распределение классов в обучающей выборке:")
print(y_train.value_counts(normalize=True))
```

```python
# Пайплайн 1: Стандартизация + PCA + Логистическая регрессия
pipe_lr = make_pipeline(
    StandardScaler(),
    PCA(n_components=2),
    LogisticRegression(random_state=my_seed, solver='lbfgs', max_iter=1000)
)

# Оценка точности с перекрёстной проверкой
print("Оценка точности логистической регрессии...")
cv_scores_lr = cross_val_score(estimator=pipe_lr, X=X_train, y=y_train, cv=5, scoring='accuracy')
print(f'Acc с перекрёстной проверкой для pipe_lr: {np.mean(cv_scores_lr):.3f} (+/- {np.std(cv_scores_lr):.3f})')
```

```python
# Пайплайн 2: Стандартизация + PCA + SVM
pipe_svc = make_pipeline(
    StandardScaler(),
    PCA(n_components=2),
    SVC(random_state=my_seed)
)

# Упрощаем параметры сеточного поиска
param_range_fast = [0.01, 0.1, 1.0, 10.0]  # меньше значений
param_grid_fast = [
    {'svc__C': [0.1, 1.0], 'svc__kernel': ['linear']},
    {'svc__C': [0.1, 1.0], 'svc__gamma': [0.01, 0.1], 'svc__kernel': ['rbf']}
]

# Уменьшаем количество разбиений CV
kfold_fast = KFold(n_splits=3, random_state=my_seed, shuffle=True)

gs_fast = GridSearchCV(
    estimator=pipe_svc,
    param_grid=param_grid_fast,
    scoring='accuracy',
    cv=kfold_fast,
    n_jobs=-1,
    verbose=1
)
```

```python
print("Запуск быстрого сеточного поиска для SVM...")
tic = time.perf_counter()
gs_fast.fit(X_train, y_train)
toc = time.perf_counter()
print(f"Быстрый сеточный поиск занял {toc - tic:.2f} секунд")
```

```python
# Точность лучшей модели SVM
print(f'Acc лучшей модели SVM: {gs_fast.best_score_:.3f}')
print(f'Параметры лучшей модели: {gs_fast.best_params_}')
```

```python
# Сравнение моделей
models = ['sc_pca_logit', 'sc_pca_svc']
scores = [np.mean(cv_scores_lr), gs_fast.best_score_]
comparison = pd.DataFrame({'Модель': models, 'Acc': scores})
print("Сравнение моделей:")
print(comparison)
```

```python
# Прогноз на отложенные наблюдения с помощью лучшей модели
best_model = gs_fast.best_estimator_
y_hat = best_model.predict(X_predict)

# Точность прогноза
print("Отчёт о классификации на отложенной выборке:")
print(classification_report(y_predict, y_hat))
```

```python
# Если точность ниже 96%, попробуем увеличить количество компонент PCA
if gs_fast.best_score_ < 0.96:
    print("\nПопытка улучшить точность с помощью большего количества компонент PCA...")
    
    best_pca_score = gs_fast.best_score_
    best_n_components = 2
    
    # Используем лучшие параметры из быстрого поиска
    best_svc_params = {k.replace('svc__', ''): v for k, v in gs_fast.best_params_.items() if k.startswith('svc__')}
    
    # Определяем максимальное возможное количество компонент
    max_components = min(X_train.shape[0], X_train.shape[1])
    print(f"Максимальное возможное количество компонент PCA: {max_components}")
    
    # Создаем список компонент, не превышающий максимум
    possible_components = [n for n in [5, 10, 15, 20] if n <= max_components]
    if not possible_components:  # если все значения слишком большие
        possible_components = list(range(2, min(8, max_components + 1)))
    
    print(f"Будем проверять компоненты: {possible_components}")
    
    for n_components in possible_components:
        try:
            pipe_svc_improved = make_pipeline(
                StandardScaler(),
                PCA(n_components=n_components),
                SVC(**best_svc_params, random_state=my_seed)
            )
            
            # Используем меньше разбиений для ускорения
            cv_scores = cross_val_score(pipe_svc_improved, X_train, y_train, cv=3, scoring='accuracy')
            mean_score = np.mean(cv_scores)
            print(f"PCA components: {n_components}, Acc: {mean_score:.3f}")
            
            if mean_score > best_pca_score:
                best_pca_score = mean_score
                best_n_components = n_components
                
        except ValueError as e:
            print(f"Ошибка при n_components={n_components}: {e}")
            continue
    
    print(f"\nЛучшее количество компонент PCA: {best_n_components}")
    print(f"Лучшая точность: {best_pca_score:.3f}")
    
    if best_pca_score >= 0.96:
        print("Требование достигнуто: точность >= 96%")
        
        # Переобучим лучшую модель с оптимальным количеством компонент
        best_pipe = make_pipeline(
            StandardScaler(),
            PCA(n_components=best_n_components),
            SVC(**best_svc_params, random_state=my_seed)
        )
        
        best_pipe.fit(X_train, y_train)
        y_hat_improved = best_pipe.predict(X_predict)
        
        print("\nУлучшенный отчёт о классификации на отложенной выборке:")
        print(classification_report(y_predict, y_hat_improved))
        
        # Обновляем лучшую модель
        best_model = best_pipe
    else:
        print("Не удалось достичь точности 96%")
        print("Используем лучшую модель из первоначального поиска")
        best_model = gs_fast.best_estimator_
else:
    print("Требование уже выполнено: точность >= 96%")
    best_model = gs_fast.best_estimator_
```
